```python
%matplotlib inline
import numpy as np
import healpy as hp
import matplotlib.pylab as plt
import lsst.sims.featureScheduler as fs
from lsst.sims.utils import _hpid2RaDec
```

```python
nside = 128
current_obs = np.random.rand(hp.nside2npix(nside))
ra, dec = _hpid2RaDec(nside, np.arange(hp.nside2npix(nside)))
out =  np.where( (ra < np.radians(45)) | (ra > np.radians(55.)))
current_obs[out] = hp.UNSEEN
current_obs[np.where(np.abs(dec) > np.radians(7.))] = hp.UNSEEN
```

```python
hp.mollview(current_obs)
```

```python
hpc = fs.hpmap_cross(nside=nside)
```

```python
print hpc(current_obs, np.radians(50), np.radians(0.), 0.)
print hpc(current_obs, np.radians(50), np.radians(0.), np.radians(30.))
```

```python
ra_point, dec_point, obs_map = hpc(current_obs, np.radians(50), np.radians(0.), np.radians(30.), return_pointings_map=True)
hp.mollview(obs_map)
good = np.where(current_obs != hp.UNSEEN)
obs_map[good] = 3
hp.mollview(obs_map)
```
